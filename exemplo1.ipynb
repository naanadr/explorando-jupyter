```python
import pandas as pd

url = (
  "https://pt.wikipedia.org/wiki/Lista_de_parlamentares_de_Pernambuco"
)

tables = pd.read_html(url)


deputados = tables[1]

deputados = deputados[:-1]

mais_eleito = deputados[deputados.Mandatos == deputados.Mandatos.max()]
nome_mais_eleito = mais_eleito["Deputados federais eleitos"].loc[0]
```

```python
deputados_municipio = deputados.groupby("Naturalidade", as_index=False).agg(Qtd_Deputados=("Mandatos", "count")).sort_values(by='Qtd_Deputados', ascending=False)
deputados_municipio.to_csv("deputados_por_municio_pe.csv")

deputados_municipio = deputados_municipio[:10]

deputados_municipio.plot.barh(
    x="Naturalidade", 
    y="Qtd_Deputados", 
    title="Naturalidade Deputados Pernambuco", 
    xlabel="Quantidade de deputados", 
    legend=False
).invert_yaxis()
```

```python
deputados_ = deputados.sort_values(by='Mandatos', ascending=False)

deputados_ = deputados_[:10].sort_values(by='Mandatos', ascending=True)
deputados_["Mandatos"] = deputados_["Mandatos"].astype(int)

deputados_.plot.barh(
    x="Deputados federais eleitos", 
    y="Mandatos", 
    xlabel="Quantidade de mandatos", 
    ylabel="Nome deputado",
    title="Deputados mais eleitos em Pernambuco",
    legend=False
)
```
